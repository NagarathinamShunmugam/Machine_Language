```python
import pandas as pd #pandas is a library user for data set manipulation
```

```python
dataset=pd.read_csv("salaries.csv")#read input file using pandas read function
```

```python
dataset#display stored dataset in the variable
```

```python
Independent=dataset[["years_of_experience"]]#split input from dataset
```

```python
dependent=dataset[["salary"]]#split ouput from dataset
```

```python
Independent#display input dataset
```

```python
dependent#display ouput dataset
```

```python
from sklearn.model_selection import train_test_split#import split test and traing dataset function using sklearn library
```

```python
train_test_split(Independent,dependent,test_size=0.30,random_state=0)#split test and traing dataset using function
```

```python
x_train,x_test,y_train,y_test=train_test_split(Independent,dependent,test_size=0.30,random_state=0)#assign splitted test and train dataset to diff variables
```

```python
x_train#display input training dataset
```

```python
y_test#display output test dataset
```

```python
#model creation
from sklearn.linear_model import LinearRegression#import linearRegression class
regressor=LinearRegression()#assign class to variable
regressor.fit(x_train,y_train)#training the model using train dataset
```

```python
#display the weight
weight=regressor.coef_
weight
```

```python
#display the bias
bias=regressor.intercept_
bias
```

```python
#test the created model
y_pred=regressor.predict(x_test)#assign the predicted output to variable
```

```python
from sklearn.metrics import r2_score#Evaluate the model using R2
```

```python
r_score=r2_score(y_test,y_pred)#passing arguments as predited and actual output r2 function for Evaluatation
r_score#display r2 value
```

```python
#save the best model
import pickle#library to convert object to bites and save that in file
filename="finalized_linear.sav"#create a empty file
```

```python
pickle.dump(regressor,open(filename,'wb'))#write all functions of regressor into newly created file
```

```python
#load the saved model
loaded_model=pickle.load(open(filename,'rb'))
result=loaded_model.predict([[17]])
```
