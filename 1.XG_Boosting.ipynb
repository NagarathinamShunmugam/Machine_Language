```python
import pandas as pd
```

```python
import xgboost
```

```python
dataset=pd.read_csv("insurance_pre.csv")
```

```python
dataset
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)
```

```python
dataset=dataset.astype(int)
```

```python
dataset
```

```python
dataset.columns
```

```python
independent=dataset[['age', 'bmi', 'children', 'sex_male', 'smoker_yes']]
```

```python
independent
```

```python
dependent=dataset[['charges']]
```

```python
dependent
```

```python
 from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0)#assign splitted test and train dataset to diff variables
```

```python
x_train
```

```python
from xgboost import XGBRegressor
```

```python
XGB=XGBRegressor(max_depth=3, learning_rate=0.01, n_estimators=500, objective='reg:linear', booster='gbtree')
```

```python
XGB.fit(x_train,y_train)
```

```python
y_pred=XGB.predict(x_test)
```

```python
from sklearn.metrics import r2_score
```

```python
r_score=r2_score(y_test,y_pred)#passing arguments as predited and actual output r2 function for Evaluatation
r_score
```

```python
#save the best model
import pickle#library to convert object to bites and save that in file
filename="finalized_model_XG_Boosting_Regression.sav"#create a empty file
pickle.dump(XGB,open(filename,'wb'))#write all functions of regressor into newly created file
```

```python
#load the saved model
loaded_model=pickle.load(open(filename,'rb'))
result=loaded_model.predict([[34,33,0,1,0]])
result
```
